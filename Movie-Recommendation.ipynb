```python
# Define a function to clean the 'Gross' column
def clean_gross(value):
    if isinstance(value, str):
        # Remove the dollar sign
        value = value.replace('$', '')
        
        # Check for 'M' (millions) or 'K' (thousands) and convert accordingly
        if 'M' in value:
            value = float(value.replace('M', '')) * 1e6
        elif 'K' in value:
            value = float(value.replace('K', '')) * 1e3
        else:
            value = float(value)
    
    return value

# Apply the cleaning function to the 'Gross' column
df['Gross'] = df['Gross'].apply(clean_gross)

# Check the cleaned data
df[['Gross']].head()
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
# Initialize the k-NN model
knn_model = NearestNeighbors(n_neighbors=5, algorithm='auto')

# Train the model using the normalized feature set
knn_model.fit(features_scaled)
```

```python
def recommend_movies(movie_index, model, features, df, n_recommendations=5):
    # Get the distances and indices of the nearest neighbors
    distances, indices = model.kneighbors([features[movie_index]])

    # Get the names of the recommended movies
    recommended_movies = []
    for idx in indices[0][1:n_recommendations+1]:  # Skip the first one (it's the same movie)
        recommended_movies.append(df.iloc[idx]['Movie Name'])

    return recommended_movies
```

```python
# Let's say we want to find movies similar to the one at index 0
movie_index = 2
similar_movies = recommend_movies(movie_index, knn_model, features_scaled, df, n_recommendations=5)

# Print the results
print(f"Movies similar to {df.iloc[movie_index]['Movie Name']}:")
for movie in similar_movies:
    print(movie)
```
